```python
from pathlib import Path
import pandas as pd
```

```python
nn_mapes = Path.home() / 'OneDrive' / 'Kirstin' / 'Uni' / 'Year4' / 'MSciProject' / 'SummerWrite-Up' / 'nn_mapes.csv'
assert nn_mapes.exists()

rf_mapes = Path.home() / 'OneDrive' / 'Kirstin' / 'Uni' / 'Year4' / 'MSciProject' / 'SummerWrite-Up' / 'rf_mapes.csv'
assert rf_mapes.exists()
```

```python
nn_data = pd.read_csv(nn_mapes)
nn_data.set_index('site', inplace=True)

rf_data = pd.read_excel(rf_mapes)
rf_data.set_index('site', inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Create the heatmap
fig, ax = plt.subplots(figsize=(12, 8))
sns.set_theme(style='ticks', font='Arial')


sns.heatmap(nn_data, 
            annot=True, 
            fmt=".2f",
            annot_kws={"fontsize": 10},
            cmap='RdYlGn_r', 
            linewidths=0.5, 
            linecolor='white', 
            norm = mcolors.LogNorm(vmin=nn_data.min().min(), vmax=nn_data.max().max())
            )

# y-axis
plt.yticks(rotation=0)
plt.ylabel('AGAGE site code', fontstyle='italic', fontsize=12)

# x-axis
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False, 
                labeltop=True, size=10)
plt.tick_params(axis='x', which='both', top=True, size=5)

ax.xaxis.set_label_position('top')
plt.xlabel('Trace species', fontstyle='italic', fontsize=12)

# adding label to colour bar
cbar = ax.collections[0].colorbar
cbar.set_label('Mean Absolute Percentage Error (%)', fontstyle='italic', fontsize=12)
# putting colour bar on the right
cbar.set_label('Mean Absolute Percentage Error (%)', fontstyle='italic', fontsize=12)


plt.tight_layout()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Create the heatmap
fig, ax = plt.subplots(figsize=(12, 8))
sns.set_theme(style='ticks', font='Arial')

vmin = 0.001
vmax = 6.2

sns.heatmap(nn_data, 
            annot=True, 
            fmt=".2f",
            annot_kws={"fontsize": 14},
            cmap='viridis', 
            linewidths=0.5, 
            linecolor='white', 
            norm = mcolors.LogNorm(vmin=vmin, vmax=vmax)
            )
# y-axis
plt.yticks(rotation=0)
plt.ylabel('AGAGE site code', fontstyle='italic', fontsize=12)
plt.yticks(fontsize=12)

# x-axis
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False, 
                labeltop=True, size=14)
plt.tick_params(axis='x', which='both', top=True, size=5)
# increasing fontsize of x-axis labels
plt.xticks(fontsize=12)

ax.xaxis.set_label_position('top')
plt.xlabel('Trace species', fontstyle='italic', fontsize=12)

# adding label to colour bar
cbar = ax.collections[0].colorbar
cbar.set_label('Mean Absolute Percentage Error (%)', fontstyle='italic', fontsize=14)


plt.tight_layout()
```
